```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.cluster import KMeans

coordenadas = np.loadtxt('coordenadasPacientes.txt')

latitud = coordenadas[:, 0]
longitud = coordenadas[:, 1]

plt.scatter(latitud, longitud, s=10) 

plt.xlabel('Latitud')
plt.ylabel('Longitud')
plt.title('Ubicaciones geogrficas de los pacientes infectados')
plt.show()
```

Punto B

```python
def graph_k_means(coordenadas, centroids, labels, title):
    latitud = coordenadas[:, 0]
    longitud = coordenadas[:, 1]
    lat_centroids = centroids[:, 0]
    lon_centroids = centroids[:, 1]

    plt.figure()
    
    scatter = plt.scatter(latitud, longitud, c=labels, alpha=0.3)
    plt.scatter(lat_centroids, lon_centroids, c='red', marker='P', s=100, label='Centros de atencin')

    plt.xlabel('Latitud')
    plt.ylabel('Longitud')
    plt.title(title)

    legend_names = []

    for i, centroid in enumerate(centroids):
        legend_names.append(f'Centro {i+1}: ({centroid[0]:.3f}, {centroid[1]:.3f})')

    legend = plt.legend(handles=scatter.legend_elements()[0], labels=legend_names, loc='upper left')

    for legendHandle in legend.legendHandles: 
        legendHandle.set_alpha(1)
    plt.show()
```

```python
k_means = KMeans(n_clusters=4, random_state=0)
k_means.fit(coordenadas)

labels = k_means.labels_
centroids = k_means.cluster_centers_

graph_k_means(coordenadas, centroids, labels, 'Ubicaciones geogrficas de los pacientes infectados y los centros de atencin')
```

Punto C

```python
k_means2 = KMeans(n_clusters=4, random_state=42)
k_means2.fit(coordenadas)

labels2 = k_means2.labels_
centroids2 = k_means2.cluster_centers_

graph_k_means(coordenadas, centroids, labels, 'Ubicaciones geogrficas de los pacientes infectados (Semilla 0)')
graph_k_means(coordenadas, centroids2, labels2, 'Ubicaciones geogrficas de los pacientes infectados (Semilla 42)')
```
