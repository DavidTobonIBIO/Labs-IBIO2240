```python
k_means = KMeans(n_clusters=4)
k_means.fit(coordenadas)

labels = k_means.labels_
centroids = k_means.cluster_centers_

graph_k_means(coordenadas, centroids, labels, 'Ubicaciones geográficas de los pacientes infectados y los centros de atención')
```
